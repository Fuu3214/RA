```python
# prerequisites
import numpy as np
import matplotlib.pyplot as plt
from keras.datasets import mnist
from keras.layers import Input, Dense, Lambda
from keras.models import Model
from keras import backend as K
from keras import objectives
from scipy.stats import norm
```

```python
def enc_fc(input_shape, n_hidden, z_dim):
    x = Input(shape=input_shape)
    x_encoded = Dense(n_hidden, activation='relu')(x)
    x_encoded = Dense(n_hidden//2, activation='relu')(x_encoded)

    mu = Dense(z_dim)(x_encoded)
    log_var = Dense(z_dim)(x_encoded)  
    enc = Model(x, [mu, log_var, x_encoded], name='encoder')
    return enc
```

```python
# sampling function
def sampling(args):
    mu, log_var = args
    eps = K.random_normal(shape=(batch_size, z_dim), mean=0., stddev=1.0)
    return mu + K.exp(log_var) * eps
```

```python
# decoder
def dec_fc(n_hidden, x_shape):
    z_latent = Input(shape=(z_dim,), name='z_sampling')
    z_decoded = Dense(n_hidden//2, activation='relu')(z_latent)
    z_decoded = Dense(n_hidden, activation='relu')(z_decoded)
    x_hat = Dense(x_shape, activation='sigmoid')(z_decoded)
    dec = Model(z_latent, x_hat, name='decoder')
    return  dec
```

```python
def classify_layer(x_encoded, y_dim):
    logits = Dense(32, activation='relu')(x_encoded)
    y_prob = Dense(y_dim, activation='softmax')(logits)
    cl = Model(x_encoded, y_prob, name='classify_layer') 
    return cl
```

```python
def descriminator(y_dim, z_dim, n_hidden):
    y = Input(shape=(y_dim,), name='y')
    z_y = Dense(n_hidden//2, activation='relu')(y)
    z_y = Dense(n_hidden, activation='relu')(z_y)

    z_y_mu = Dense(z_dim)(z_y)
    z_y_log_var = Dense(z_dim)(z_y)
    des = Model(y, [z_y_mu, z_y_log_var], name='encoder')
    return des
```

```python
# loss
def vae_loss(x, x_hat, x_shape, mu, log_var):
    reconstruction_loss = objectives.binary_crossentropy(x, x_hat) * x_shape
    kl_loss = 0.5 * K.sum(K.square(mu) + K.exp(log_var) - log_var - 1, axis = -1)
    return reconstruction_loss + kl_loss
```

```python
def vae(enc, dec, args):
    input_shape, x_shape, n_hidden, z_dim = args
    x = Input(shape=input_shape)
    mu, log_var, x_encoded = enc(x)
    z_sampled = Lambda(sampling, output_shape=(z_dim,))([mu, log_var])
    x_hat = dec(z_sampled)

    vae = Model(x, x_hat)
    vae.add_loss(vae_loss(x, x_hat, x_shape, mu, log_var))
    vae.compile(optimizer='rmsprop')
    return vae
```

```python
# data load
(x_tr, y_tr), (x_te, y_te) = mnist.load_data()
x_tr, x_te = x_tr.astype('float32')/255., x_te.astype('float32')/255.
x_tr, x_te = x_tr.reshape(x_tr.shape[0], -1), x_te.reshape(x_te.shape[0], -1)
print(x_tr.shape, x_te.shape)

# network parameters
batch_size, n_epoch = 100, 100
n_hidden, z_dim = 256, 2
x_shape = x_tr.shape[1]
input_shape = x_tr.shape[1:]
```

```python
print("ENCODER:")
enc = enc_fc(input_shape, n_hidden, z_dim)
enc.summary()
print("\n")
print("DECODER:")
dec = dec_fc(n_hidden, x_shape)
dec.summary()
print("\n")
print("VAE:")
args = [input_shape, x_shape, n_hidden, z_dim]
vae = vae(enc,dec,args)
vae.summary()
```

```python
# train
vae.fit(x_tr,
       shuffle=True,
       epochs=n_epoch,
       batch_size=batch_size,
       validation_data=(x_te, None), verbose=1)
```

```python
# Plot of the digit classes in the latent space
x_te_latent = enc.predict(x_te, batch_size=batch_size)[0]
plt.figure(figsize=(6, 6))
plt.scatter(x_te_latent[:, 0], x_te_latent[:, 1], c=y_te)
plt.colorbar()
plt.show()
```

```python
# display a 2D manifold of the digits
n = 15 # figure with 15x15 digits
digit_size = 28
figure = np.zeros((digit_size * n, digit_size * n))

grid_x = norm.ppf(np.linspace(0.05, 0.95, n)) 
grid_y = norm.ppf(np.linspace(0.05, 0.95, n))

for i, yi in enumerate(grid_x):
    for j, xi in enumerate(grid_y):
        z_sample = np.array([[xi, yi]])
        x_decoded = dec.predict(z_sample)
        digit = x_decoded[0].reshape(digit_size, digit_size)
        figure[i * digit_size: (i + 1) * digit_size,
               j * digit_size: (j + 1) * digit_size] = digit

plt.figure(figsize=(10, 10))
plt.imshow(figure, cmap='Greys_r')
plt.show()
```
